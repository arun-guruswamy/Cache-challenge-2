```python
import numpy as np
import matplotlib.pyplot as plt

scores = np.load('ml_data/CNNVS_scores.npy')
scores = scores.astype('float64')
pharmit_scores = np.load('pharmit_CNNVSscores.npy')
pharmit_scores = pharmit_scores.astype('float64')

# evaluate the histogram
values, base = np.histogram(pharmit_scores, bins='auto')
values1, base1 = np.histogram(scores, bins='auto')
#evaluate the cumulative
cumulative = np.cumsum(values)
cumulative1 = np.cumsum(values1)
# plot the cumulative function
plt.plot(base[:-1], cumulative/len(pharmit_scores), c='blue')
#plot the survival function
plt.plot(base1[:-1], cumulative1/len(scores), c='green')

plt.show()
```
